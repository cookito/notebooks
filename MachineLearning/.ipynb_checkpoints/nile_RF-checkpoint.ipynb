```python
%matplotlib inline
import statsmodels as sm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.ensemble import RandomForestRegressor

def load_data():
    df_read = sm.datasets.nile.load_pandas().data
    s_date = pd.Series(
        [pd.to_datetime(str(int(y_str))) for y_str in df_read['year']]
    )
    df = df_read.set_index(s_date)
    df = df.drop('year', axis=1)

    return df

df_nile = load_data()

# Plot Time Series Data
plt.figure()
ax = df_nile['volume'].plot(figsize=(8,4), grid=True)
ax.set_xlabel('year')
ax.set_ylabel('volume x 10^8 m^3')
plt.show()

df_nile['lag1'] = df_nile['volume'].shift(1) 
df_nile['lag2'] = df_nile['volume'].shift(2)
df_nile['lag3'] = df_nile['volume'].shift(3)

df_nile = df_nile.dropna()

x_train = df_nile[['lag1', 'lag2', 'lag3']][:67].values
x_test = df_nile[['lag1', 'lag2', 'lag3']][67:].values

y_train = df_nile['volume'][:67].values
y_test = df_nile['volume'][67:].values

# bootstrap:Bootstrapするかどうか
# n_estimators:木の本数
# criterion:評価関数 
# max_features:説明変数の数(推奨値は分類：√n 回帰：n/3 特徴量がいっぱいあるときはもっと増やしたほうが良い) 
# max_leaf_nodes:最大ノード数（推奨値は分類：１　回帰５）
# random_state:乱数のシード
# oob_score: R^2をout of bags使って評価するかどうか
rf = RandomForestRegressor(
    bootstrap=True,
    n_estimators=100,
    criterion='mse', 
    max_features='auto',
    random_state=1,
    oob_score=True,
)


rf.fit(x_train, y_train)
train_pred = rf.predict(x_train)
test_pred = rf.predict(x_test)

pred = np.concatenate((train_pred, test_pred))
df_nile['pred'] = pred

plt.figure()
df_nile[['volume', 'pred']].plot()
print(rf.oob_score_)

# 実はテスト用のデータの特徴量に、訓練データの特徴量が時系列でスライドしてる分混じってしまっている
```

```python
# parameter tuning by grid search
from sklearn.grid_search import GridSearchCV 

parameters = {
    'n_estimators'           : [10, 50, 100, 150, 200, 250, 300],
    'max_features'          : [1, 2, 3],
    'random_state'         : [1],
}

rf_tuned = GridSearchCV(RandomForestRegressor(), parameters)
rf_tuned.fit(x_train, y_train)

# クロスバリデーションによって得られた最適なパラメータの表示
print(rf_tuned.best_estimator_)
```
