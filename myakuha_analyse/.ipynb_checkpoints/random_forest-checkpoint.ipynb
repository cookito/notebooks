```python
%matplotlib inline
import statsmodels as sm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.grid_search import GridSearchCV

train_df = pd.read_csv('train.tsv', index_col='id', delimiter='\t')
train_df
```

```python
x_train = train_df.ix[:, :'V29999']
y_train = train_df.ix[:, 'Y']
```

```python
y_train.describe()
```

```python
time = np.arange(30000)
pulse = x_train.ix[0, :].values.reshape(30000)
print()
plt.figure()
plt.plot(time, pulse)
```

```python
x_train.ix[0, :].describe(percentiles=[0.05, 0.10, 0.25, 0.50, 0.75, 0.90, 0.95])

x_train.ix[163, :].plot(kind='hist')
```

```python
rf = RandomForestRegressor(
    bootstrap=True,
    n_estimators=400,
    criterion='mse', 
    max_features='sqrt',
    random_state=1,
    oob_score=True,
)

rf.fit(x_train, y_train)
rf.oob_score_
```

```python
parameters = {
    'n_estimators'           : [350, 400, 500],
    'max_features'          : [100],
    'random_state'         : [1],
}

rf_tuned = GridSearchCV(RandomForestRegressor(), parameters)
rf_tuned.fit(x_train, y_train)

# クロスバリデーションによって得られた最適なパラメータの表示
print(rf_tuned.best_estimator_)
```

```python
x_test = pd.read_csv('test.tsv', index_col='id', delimiter='\t')
```

```python
pred = np.ones(57)*y_train.median()
y_test = pd.DataFrame({
    'id': x_test.index,
    'pred': pred})
y_test.to_csv('median.csv', header=None, index=None)
```

```python
median_pulse_train = x_train.median(axis=1)
median_pulse_train.values
```

```python
median_pulse_test = x_test.median(axis=1)
median_pulse_test.ix[188] = 1.541
```

```python
time = np.arange(100)
pulse = x_test.ix[188, :100].values.reshape(100)
plt.figure()
plt.plot(time, pulse)
```

```python
x_test.ix[188, :].describe()
```

```python
from sklearn.linear_model import Ridge

parameters = {
    'alpha' : np.arange(0.030, 0.039, 0.0001),
}

ridge_tuned = GridSearchCV(Ridge(), parameters)
ridge_tuned.fit(median_pulse_train.reshape(144, 1), y_train.reshape(144, 1))

# クロスバリデーションによって得られた最適なパラメータの表示
print(ridge_tuned.best_estimator_)
```

```python
ridge_tuned = Ridge(0.0369)
ridge_tuned.fit(median_pulse_train.reshape(144, 1), y_train.reshape(144, 1))
pred = ridge_tuned.predict(median_pulse_test.reshape(-1, 1))

y_test = pd.DataFrame({
    'id': x_test.index,
    'pred': pred.reshape(-1)})
y_test.to_csv('ridge_rigression_by_pulse_median.csv', header=None, index=None)
```

```python
y_test['pred'].describe()
```

```python
time = np.arange(y_test['id'].count())
pulse = y_test['pred'].reshape(y_test['id'].count())
plt.figure()
plt.plot(time, pulse)
```
